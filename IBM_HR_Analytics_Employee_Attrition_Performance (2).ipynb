## dependent and independent variable 

```python
X = df_encoded.drop("Attrition",axis=1)
Y = df_encoded["Attrition"]
```

```python
##  Train The Model
```

```python
# creating train test data 
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(X,Y,test_size=0.3,random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape 
```

```python
# model building 
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model
```

```python
#model fitting
model.fit(x_train,y_train)
```

```python
# Predict on test data
y_pred = model.predict(x_test)
y_pred_proba = model.predict_proba(x_test)[:,1]
y_pred
```

```python
# Evaluation
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nAccuracy Score:", accuracy_score(y_test, y_pred))
```

```python
# Compute ROC curve and AUC
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)

# Plot the ROC Curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')  # Diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve - Log Reg')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python

fpr, tpr, thresholds=roc_curve(y_test,y_pred)
for i in thresholds:
    print(i)
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, roc_curve,precision_score

fpr, tpr, thresholds = roc_curve(y_test, y_pred)

# Initialize accuracy list
acc = []
prec=[]

# Loop over thresholds and calculate accuracy
for thres in thresholds:
    y_pred_new = np.where(y_pred > thres, 1, 0)
    acc.append(accuracy_score(y_test, y_pred_new))
    prec.append(precision_score(y_test, y_pred_new))  

# Create result dataframe
result = pd.DataFrame({'threshold': thresholds, 'accuracy': acc,"Precison":prec})

# Sort by accuracy
result_sorted = result.sort_values(by="Precison", ascending=False)

# Display top results
print(result_sorted.head())

# by using this threshold we can get more accuracy choose threshold according to accuracy
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

#model fitting 
rf= RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
# model prediction 
rf_pred =rf.predict(x_test)
rf_proba = rf.predict_proba(x_test)[:,1]
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def evaluate_model(y_test, y_pred, y_proba):
    return {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1 Score': f1_score(y_test, y_pred),
        'ROC AUC': roc_auc_score(y_test, y_proba)
    }

results = {
    'Logistic Regression': evaluate_model(y_test, y_pred, y_pred_proba),
    'Decision Tree': evaluate_model(y_test, dt_pred, dt_proba),
    'Random Forest': evaluate_model(y_test, rf_pred, rf_proba)
}

scorecard = pd.DataFrame(results).T
print(scorecard)
```

## MODEL TUNING

```python
### Logistic Regression
```

```python
from sklearn.model_selection import GridSearchCV

param_lr = {'C': [0.01, 0.1, 1, 10]}
grid_lr = GridSearchCV(LogisticRegression(max_iter=1000), param_lr, cv=5)
grid_lr.fit(x_train, y_train)
best_lr = grid_lr.best_estimator_
```

```python
### Decision Tree
```

```python
param_dt = {'max_depth': [3, 5, 10, None], 'min_samples_split': [2, 5, 10]}
grid_dt = GridSearchCV(DecisionTreeClassifier(), param_dt, cv=5)
grid_dt.fit(x_train, y_train)
best_dt = grid_dt.best_estimator_
```

### Random Forest

```python
param_rf = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}
grid_rf = GridSearchCV(RandomForestClassifier(), param_rf, cv=3)
grid_rf.fit(x_train, y_train)
best_rf = grid_rf.best_estimator_
```

### Re-Evaluate with Tuned Models

```python
best_results = {
    'Logistic Regression (Tuned)': evaluate_model(y_test, best_lr.predict(x_test), best_lr.predict_proba(x_test)[:, 1]),
    'Decision Tree (Tuned)': evaluate_model(y_test, best_dt.predict(x_test), best_dt.predict_proba(x_test)[:, 1]),
    'Random Forest (Tuned)': evaluate_model(y_test, best_rf.predict(x_test), best_rf.predict_proba(x_test)[:, 1])
}

tuned_scorecard = pd.DataFrame(best_results).T
print(tuned_scorecard)
```

## we should prefer Random Forest(tuned) model as its accuracy is best with best ROC-AUC score
